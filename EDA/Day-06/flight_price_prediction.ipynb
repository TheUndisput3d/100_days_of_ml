```python
def cnv(el):
  lst = el.split()
  if len(lst) > 1:
    return int(lst[0])*60+int(lst[1])
  else:
    return int(lst[0])*60
```

```python
df['Duration'] = df['Duration'].str.replace(r"[hm]", "", regex=True).apply(cnv)
```

```python
from sklearn.model_selection import train_test_split
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(train_X, train_y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
df.Additional_Info.unique()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
df.Total_Stops.unique()
```

```python
add_info_ord = ['In-flight meal not included',
            'No check-in baggage included',
            'Red-eye flight',
            '1 Short layover',
            '1 Long layover',
            '2 Long layover',
            'Change airports',
            'Business class']
tot_stops_ord = ['non-stop', '1 stop', '2 stops', '3 stops', '4 stops']
```

```python
oe = OrdinalEncoder(
    categories=[add_info_ord, tot_stops_ord])
```

```python
ohe = OneHotEncoder(
    sparse_output=False,
    drop='first',
    handle_unknown='ignore'
)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
transformer = ColumnTransformer(
    transformers= [
        ('trf1', oe, ['Additional_Info', 'Total_Stops']),
        ('trf2', ohe, ['Airline', 'Source', 'Destination']),
        ('trf3', scaler, ['Duration', 'Day', 'Month', 'Year', 'Arrival_hour',
       'Arrival_minute', 'Dep_hour', 'Dep_minute'])

    ],
    remainder = 'passthrough'
)
X_transformed = transformer.fit_transform(X_train_t)
```

```python
pd.DataFrame(X_transformed)
```

```python
X_test_transf = transformer.transform(X_test_t)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_transformed, y_train_t)
```

```python
y_pred = lr.predict(X_test_transf)
```

```python
opdf = pd.DataFrame(y_pred, columns=['y_pred'])
pd.concat([opdf,pd.DataFrame(y_test_t)], axis=1)
```

```python
from sklearn.metrics import root_mean_squared_error
rmse = root_mean_squared_error(y_test_t, y_pred)
print(f"rmse: {rmse}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor

rf = DecisionTreeRegressor()
rf.fit(X_transformed, y_train_t)
y_pred_rf = rf.predict(X_test_transf)
```

```python
rmse = root_mean_squared_error(y_test_t, y_pred_rf)
print(f"rmse: {rmse}")
```
